<h1>Generate Training Samples for Azure Wide and Deep Recommender.</h1>

```python
import pandas as pd
from random import randrange
```

```python
pd.set_option('max_rows', 2000)
pd.set_option('max_columns', 100)
pd.set_option('max_colwidth', 100)
import pandas as pd
desired_width = 400    
pd.set_option('display.width', desired_width)
```

```python
#Create Rating Table
d       = {}
n_users = 10
n_rest  = 20
d['user']       = []
d['restaurant'] = []
d['rating']     = []
for u in range(1,n_users+1):
    for re in range(1, n_rest+1):
        #Remove some of the ratings so that we can then predict them:
        rating = 1 if re%2 != u%2 else 6
        if randrange(4) != 3:
            d['user'].append(u)
            d['restaurant'].append(re)
            d['rating'].append(rating)
        else:
            print('Removed User:', u, 'for restaurant', re, 'rating:', rating)
df_rating = pd.DataFrame(d)    
df_rating.iloc[0:100]
```

```python
#Create a user table:
d            = {}
d['user']    = []
d['style']   = []
d['name']    = []
for u in range(1,n_users+1):
    d['user'].append(u)
    if u%2 == 0:
        d['style'].append('French')
        d['name'].append('Monsieur_'+str(u))
    else:
        d['style'].append('Italian')
        d['name'].append('Signora_'+str(u))
df_user = pd.DataFrame(d)    
df_user
```

```python
#Create Restaurant Table:
d                  = {}
d['restaurant']    = []
d['style']         = []
d['name']          = []
for re in range(1,n_rest+1):
    d['restaurant'].append(re)
    if re%2 == 0:
        d['style'].append('French')
        d['name'].append('Creperie_'+str(re))       
    else:
        d['style'].append('Italian')
        d['name'].append('Pizeria_'+str(re))       
df_restaurant = pd.DataFrame(d)    
df_restaurant
```

### Save the tables.

```python
df_rating.to_csv('mulx_rating_artificial_1.csv', sep=';', index=False)
df_user.to_csv('mulx_user_artificial_1.csv', sep=';', index=False)
df_restaurant.to_csv('mulx_restaurant_artificial_1.csv', sep=';', index=False)
```
